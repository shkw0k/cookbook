##What is in the DOM object?

A DOM has one and only documentElement, the root or the entire document.

```python
docRoot = dom.documentElement
```

The root element is an Element. It has a tag name, attributes and children.

```python
docRoot.tagName
```

Attributes have name and value.

```python
for x in docRoot.attributes.values():
    print ("name= %s, value= %s" % (x.name, x.value))
```

An Element has children nodes, *childNodes*. 

```python
docRoot.childNodes
```

Now let's write a function that recursively inspects what is in the entire document.

```python
def deepShow (elem, level=0):
    spaces = "   " * level
    print("\n%s" % spaces, elem.tagName, end=' ')
    for item in elem.attributes.items():
        print ("%s='%s'" % item, end=' ')
    for e in elem.childNodes:
        if e.nodeType == 3: # text
            cont = e.nodeValue.strip()
            if len(cont) > 0:
                print (cont, end=' ')
        elif e.nodeType == 1: # Element
            deepShow(e, level=level+1)
        else:
            print ("Unknown/Not implemented", e.nodeType, e.nodeName)
```

OK, try it with the document root element.

```python
deepShow(docRoot)
```

###Read a VoTable and convert it to a DataFrame.
The xml example file is actually a votable, which is a xml file with a special structure.
<br>Here we are going to extract the name of the columns and the table data and build a panda dataframe.

First we write a function that returns the element by path
instead of by tag name.

```python
def getElementsByPath (elem, path):
    """
    Given a path, ie path1/path2/path3/../pathN
    returns the elements that match the path.
    No wild char allowed.
    """
    parts = path.split('/')
    out = []
    list1 = [elem]
    for tag in parts:
        list2 = []
        for elem1 in list1:
            for child in elem1.childNodes:
                if child.nodeType == 1 and child.tagName == tag:
                    list2.append (child)
        list1 = list2
    return list1
```

Then we write a function that returns the rows of the votable. It also tries to convert the content to integers or floats, if possible.

```python
def getRows(voRoot):
    rows = []
    for tr in getElementsByPath(voRoot, "RESOURCE/TABLE/DATA/TABLEDATA/TR"):
        row = []
        for col in tr.getElementsByTagName ("TD"):
            cont = col.childNodes[0].nodeValue if col.nodeType == 1 and col.childNodes.length > 0 else "N/A"
            try:
                if len(cont) > 0:
                    intNumber = int(cont)
                    cont = intNumber
            except:
                try:
                    floatNumber = float(cont)
                    cont = floatNumber
                except:
                    pass
            row.append (cont)
        rows.append (row)
    return rows
```

Next, we write a function that returns the column names. We check for "ID" in upper and lower case because votables producers are not always consistent.

```python
def getColumns(voRoot):
    columns = []
    for e in getElementsByPath(voRoot, "RESOURCE/TABLE/FIELD"):
        for n in ["ID", "id", "Id"]:
            attr = e.getAttribute("ID")
            if attr:
                break;
        columns.append (attr)
    return columns
```

To build a pandas dataframe, we simply call DataFrame() and assign the columns.

```python
def vot2Pandas(fname):
    votable = parse(fname)
    voRoot = votable.documentElement
    df = pd.DataFrame(getRows(voRoot))
    df.columns = getColumns (voRoot)
    return df
```

Let's try it and then display the votable as HTML.

```python
df = vot2Pandas("xml/GSC23.xml")
HTML(df.to_html())
```
