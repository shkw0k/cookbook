```python
%matplotlib inline
import pandas as pd
from xml.dom.minidom import parseString
from IPython.display import display_html, HTML
```

```python
def readXML (fname):
    with open(fname, "r") as fh:
        data = fh.read()
    return parseString(data)
```

```python
dom = readXML ("NOMAD.vot")
```
