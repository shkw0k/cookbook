#How to make a photo look like a drawing

```python
import matplotlib.pyplot as plt
import numpy as np
```

#First a few useful functions

rgb2bw converts RGB image to monochromatic image. Image is numpy matrix [rows, cols, 3].

```python
def rgb2bw(rgb):
    r = rgb[:,:,0] * 0.3
    g = rgb[:,:,1] * 0.59 
    b = rgb[:,:,2] * 0.11
    return r + g + b
```

integralImage returns the sums the pixel values along both axes.

```python
def integralImage(img):
    return np.cumsum(img, axis=0).cumsum(axis=1)
```

avgAreaSlow calculates the average pixel values for each each centered at a rectangle of width and height equal n pixels.

```python
def avgAreaSlow(img, intImage, n):
    rows, cols = img.shape
    nimg = np.copy(img)
    h = max(1, int(n/2))
    hsqr = 4*h*h
    for r in range(h, rows-h):
        for c in range(h, cols-h):
            nimg[r,c] = (intImage[r-h,c-h]+intImage[r+h,c+h]-intImage[r-h,c+h]-intImage[r+h,c-h]) / hsqr
    return nimg

#%timeit img4 = avgAreaSlow(img, iimg, 7)
```

avgArea calculates the average pixel values for each pixel centered at a rectangle of width and height equal to n pixels. This version uses numpy and is 50 times faster than avgAreaSlow.

```python
def avgArea(img, intImage, n):
    h = max(1, int(n/2))
    h2 = h+h
    im1 = intImage[:-h2,:-h2]
    im2 = intImage[h2:,:-h2]
    im3 = intImage[:-h2,h2:]
    im4 = intImage[h2:,h2:]
    nimg = np.copy(img)
    nimg[h:-h,h:-h] = (im1 + im4 - im3 - im2) / (h2*h2)
    return nimg
```

#Now we are ready

First read the image and highlight the pixel values by applying x^1.5.

Then we calculate the integral image and use it to calculate two blurred images with parameters blur1 and blur2. 
Blur1 should be greater than blur2. The parameter t is used to combine the the two blurred images. The result image is then clipped to 0 and 255.

```python
img = rgb2bw(plt.imread("../images/backyard.jpg"))
img = np.power(img, 1.5)
```

```python
def img2Drawing (img, blur1=11, blur2=5, t=-15):
    iimg = integralImage(img)
    blur2 = max(3, blur2)
    blur1 = max(blur1, blur2+2)
    bimg1 = avgArea(img, iimg, blur1)
    bimg2 = avgArea(img, iimg, blur2)
    return np.clip(t * (bimg1 - bimg2) + bimg2, 0, 255)
```

Let's see the result.

```python
def plotImages(orgImage, imgDrawing):
    fig=plt.figure(figsize=(16,16))
    ax = plt.subplot(121)
    ax.imshow(orgImage, cmap='gray')
    ax.set_xlabel("Original")
    
    ax = plt.subplot(122)
    ax.imshow(imgDrawing, cmap='gray')
    ax.set_xlabel("'Drawing' version")

plotImages(img, img2Drawing(img, 9, 3, -15))
```
