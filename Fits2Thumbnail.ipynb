Remap rescales the pixels values linearly from (fromLo,fromHi) to (toLo,toHi). The result is converted to integers and clipped, such as the min and max values are in the range of (toLo,toHi).

```python
def remap (arr, fromLo, fromHi, toLo, toHi):
    m = (toHi - toLo) / (fromHi - fromLo)
    b = -m * fromLo + toLo
    return np.clip(np.int_(arr * m + b), toLo, toHi)
```

applyHistEq takes a 2D array and applies histrogram equalization on the data. It uses the numpy histogram function, which returns the histogram array and the edges array. The values of the histogram array are accumulated. loCut and hiCut are cut levels from 0 to 1, where 0 is the minimum and 1 is the maximum pixel value. The minimum is the first element in edges[] and the maximum is the last element of edges[]. The data is first remap from the desired cut levels to the number of histogram bins, which are used as indices in the accumulated array. The result is not rescaled to number of colors. 

```python
def applyHistEq(data, loCut=0, hiCut=1, nHist=1024, **kwds):
    flatData = data.flatten()
    histg, edges = np.histogram(flatData, bins=nHist, density=False)
    hsum = np.cumsum(histg)
    lo, hi = edges[0], edges[-1]
    fromLo = (hi-lo) * loCut + lo
    fromHi = (hi-lo) * hiCut + lo
    data1 = remap(data, fromLo, fromHi, 0, nHist-1)
    return hsum[np.int_(data1)]
```

Centroid calculates the barycenter of the input array. It returns the centroid and the standard deviation.

```python
def centroid(arr):
    l = arr.shape[0]
    ixs = np.arange(l)
    tmp = arr * ixs
    tmp1 = tmp * ixs
    sumarr = sum(arr)
    cen = sum(tmp)/sumarr
    return cen, math.sqrt(sum(tmp1)/sumarr - cen * cen)
```

Contrast limiting histrogram equalization. The function applyHEQ1 takes a 2D data array and returns the contrast limited histrogram equalized data. First the a histogram is calculated. Then the cendroid od the histogram is calculated. If the histogram shows one large peak, then the centroid points to that peak and the standard deviation shows how wide the peak is. If the histogram has more than one peak or it is flat, then the standard deviation is large. The parameter cutWidth gives how many time standard deviation should be used to calculate loCut and hiCut. If cutWidth is 1, then loCut is one standard deviation below centroid and hiCut is one standard deviation above.

The parameter tFactor is the factor for the threshold that limits the contrast. An image that has only one color or intensity level, ie all pixels have the same value, has no contrast. An image that pixel values equally distributed from min to max values has a flat histogram, and the increase of intensity levels is constant. If large percentage of pixels has the same or similar intensity, for example, an image of the sky with a few bright stars and a darker sky background, then the histogram shows high values for the sky intensities and low values for star intensities. The accumulated histogram shows large increases in the low intensity areas and small increases in the high intensity areas. This leads to high contrast in low intensity areas and low contrast in high intensity areas. The result is a noisy looking image. The contrast can be limited by setting a threshold on how much each histogram bin value can contribute to the accumulated histogram. If the parameter tFactor is one, then the maximum increase is approximately 34.1% (one standard deviation) of the number of pixels.



```python
def applyHEQ1(data, cutWidth=1, tFactor=1, nHist=65535):
    flatData = data.flatten()
    leng = len(flatData)
    histg, edges = np.histogram(flatData, bins=nHist, density=False)
    cen, cstd = centroid(histg)
    wing = cutWidth * cstd 
    loIdx = int(max(0, cen-wing))
    hiIdx = int(min(loIdx+wing+wing, nHist))
    fromLo = edges[loIdx]
    fromHi = edges[hiIdx]
    
    #print ("fromLo", fromLo, fromHi, cstd)
    data1 = remap(data, fromLo, fromHi, 0, nHist-1)
    histg, edges = np.histogram(data1.flatten(), bins=nHist, density=False)
    
    thold = leng / nHist * cstd * tFactor
    histg = np.clip(histg, 0, thold)
    hsum = np.cumsum(histg)
    data1 = hsum[np.int_(data1)]
    return data1
    
```

##Examples

* data2Plot calls appltHEQ1 and display the result image
* fits2Plot reads a fits file and display image via data2Plot

```python
def data2Plot (subPlotId, data, **kwds):
    data1 = applyHEQ1(data, **kwds)
    
    ax = plt.subplot(subPlotId)
    ax.imshow(data1, cmap='gray', origin="lower")
    
def fits2Jpeg (subPlotId, fname, **kwds):
    hdus = pyfits.open(fname)
    data = hdus[0].data
    data2Plot (subPlotId, data, **kwds)
    hdus.close() 
```

Brute force method to resize an image.

```python
def resizeImage (imgData, newWidth=128):
    r,c = imgData.shape
    skip = int(c/newWidth) + 1
    out = np.array([[0] * int(c / skip)] * int(r/skip))
    for r0, ri in enumerate (range(0, r -skip, skip)):
        for c0, ci in enumerate(range(0, c-skip, skip)):
            value = np.sum(imgData[ri:ri+skip,ci:ci+skip])
            out[r0][c0] = value 
    return out
```

##Eample

* fits2Thumbnail reads a fits file, resizes the image via given method and applies HEQ and returns the image
* save2Jpg saves image as jpg.

```python
def fits2Thumbnail(subPlotId, fname, newWidth=128, method="skip", **kwds):
    hdus = pyfits.open(fname)
    data = hdus[0].data
    r,c = data.shape
    skip = int(c/newWidth) +1
    if method == "sum":
        thumbnail = resizeImage (data, newWidth)
    else: # defautl skip
        thumbnail = applyHEQ1(data[0:r:skip,0:c:skip], **kwds)
    hdus.close()
    return thumbnail
```

#Example

* fits2HTML converts a list of fits files to jpeg thumbnails and creates a HTML page showing all the images.


```python
def fits2HTML (pageName, fileNames):
    fh = open(pageName, "w") 
    print ("""<html>
    <style>
    .img1 { margin: 10px; display:inline-block; }
    </style>
    <body>
    """, file=fh)
    for i, f in enumerate(fileNames):
        print ("processing", f)
        thumbnail = fits2Thumbnail((150+i), f, tFactor=0.1, method="skip")     
        basename = os.path.basename(f)
        jpegName = "out/%s" % basename.replace(".fits", ".jpg")
        plt.imsave(jpegName, thumbnail, cmap='gray', origin='lower')
        print ("<div class='img1'><img src='%s'><br>%s</div>" % (jpegName,basename), file=fh)
    print ("</body></html>", file=fh)
    fh.close()
    print ("done")
    
fits2HTML ("thumbnail.html", fileNames)
```
