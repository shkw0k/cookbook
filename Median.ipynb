# Median test

```python
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image as pimg
import PIL.ImageDraw as pdraw

import scipy.ndimage as ndimage

from numba import jit, i8, double, autojit
```

# Read Image

```python
orgImg = pimg.open("smarties.jpg").convert("L")

# Now draw a line across the image, sort of.
d1 = pdraw.Draw(orgImg)
d1.line((20, 20, 500, 400), fill='#ffffff')
img = np.array(orgImg)

plt.imshow(img, cmap='gray')
```

## Median Filter (using Numpy as kernel)

```python
def medianFilter(img, size=3):
    half = size // 2
    height, width = img.shape    
    out = np.copy(img)
    for r1 in range(half,height-half):
        r2 = r1 - half
        for c1 in range(half,width-half):
            c2 = c1 - half
            out[r1,c1] = np.median(img[r2:r2+size,c2:c2+size])
    return out
```

### Test and Show (Numpy as kernel)

```python
%%timeit
mimg = medianFilter(img, size=3)
```

# Numpy version

```python
%%timeit
mimg_ndimage = ndimage.filters.median_filter(img, size=3)
```

###The ndimage version is almost 200x faster.

# Numba Test

### 1D Version

```python
# The median function using the Qselect algorithm.

def mselHelper(a, leng, left, right):
    half = 0
    cnt = 0
    x = 0
    m = 0
    lidx = 0
    ridx = 0
    nth = leng // 2
    ok = 1
    while ok:
        half = leng // 2
        m = a[half]
        cnt = 0
        lidx = 0
        ridx = 0
        i = 0
        while i < leng:
            x = a[i]
            if x < m:
                left[lidx] = x
                lidx += 1
            elif x > m:
                right[ridx] = x
                ridx += 1
            else:
                cnt += 1                
            i += 1       
        if nth < lidx:
            a = left
            leng = lidx        
        elif nth < lidx + cnt:
            ok = 0
        else:
            a = right
            leng = ridx
            nth = nth - lidx - cnt
    return m

locals1 = {'half':i8, 'm':i8, 'a':i8[:], 'i':i8}
mselNBHelper = jit("i8(i8[:],i8,i8[:],i8[:])", locals=locals1, nopython=True)(mselHelper)
#mselNBHelper.inspect_types()

@autojit
def msel(a):
    size = len(a)
    left = np.copy(a)
    right = np.copy(a)
    return mselNBHelper(a, size, left, right)
```

```python
a = (np.random.random(10000001) * 1000).astype("int64")
```

```python
%%timeit
msel(a)
```

```python
%%timeit
np.median(a)
```

```python
size = len(a)
left = np.copy(a)
right = np.copy(a)
```

```python
%%timeit
mselNBHelper(a, size, left, right)
```

###the Numba version is no bad, but still not as fast as numpy.

## Numba Test - 2D Version

```python
# The 2d version using Qselect.
# For each nxn box, we copy the pixels to an array and then call mselHelper to find the median.

def medianFilter2D(img, out, height, width, size, left, right, buf):
    half = size // 2
    n2 = size * size
    r2 = 0
    c2 = 0
    r1 = half
    c1 = half
    i1 = 0
    i2 = 0    
    j1 = 0
    j2 = 0
    idx = 0
    while r1 < height-half:
        r2 = r1 - half
        c1 = half
        while c1 < width-half:
            c2 = c1 - half
            
            i1 = r2
            i2 = r2 + size
            idx = 0
            while i1 < i2:
                j1 = c2
                j2 = c2 + size
                while j1 < j2:
                    buf[idx] = img[i1,j1]
                    idx += 1
                    j1 += 1
                i1 += 1
                
            out[r1,c1] = mselNBHelper(buf, n2, left, right)
            c1 += 1            
        r1 += 1
    return out

locals2 = {'half':i8, 'n2':i8, 
           'r2':i8, 'c2':i8, 'r1':i8, 'c1':i8,
           'i1':i8, 'i2':i8, 'j1':i8, 'j2':i8, 'idx':i8}

medianFilterNB = jit("i8[:,:](i8[:,:],i8[:,:],i8,i8,i8,i8[:],i8[:],i8[:])", nopython=True, locals=locals2)(medianFilter2D)
```

```python
# Initialize the parameters
a = img.astype("int64")
out = np.zeros_like(a)
height, width = a.shape
size = 11
n2 = size * size
left = np.zeros(n2, dtype='i8')
right = np.zeros(n2, dtype='i8')
buf = np.zeros(n2, dtype='i8')
```

```python
%%timeit
mimg_test = medianFilterNB(a, out, height, width, size, left, right, buf)
```

```python
%%timeit
mimg_ndimage = ndimage.filters.median_filter(img, size=size)
```

The Numba version is almost as good as the ndimage version. May be mselHelper can be optimized a little bit more.
