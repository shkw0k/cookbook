# Various version of RGB 2 BW

```python
def rgb2bw(rgb):
    """
    Python style
    This is faster than the dot product.
    """
    return rgb[:,:,0] * 0.3 + rgb[:,:,1] * 0.59  + rgb[:,:,2] * 0.11

def rgb2bwDotProd(rgb):
    """
    Using Numpy dot product
    """
    return np.dot(rgb,[0.3, 0.59, 0.11])

def pilbw(rgb):
    """
    Using the PIL convert method
    """
    return np.array(pimg.fromarray(rgb).convert("L"))
```

```python
%%timeit 
rgb2bw(orgImg)
```

```python
%%timeit
rgb2bwDotProd(orgImg)
```

```python
%%timeit
pilbw(orgImg)
```

##So the PIL version is the fastest.

# Rebinning

```python
def rebin2(img, n=2):
    """
    Python style
    """
    rows, cols = img.shape
    out = [[ img[r:r+n,c:c+n].sum() for c in range(0,cols,n)] for r in range(0,rows,n)]
    return np.array(out) / (n*n)

def rebin2a(img, n=2):
    """
    Using Numpy reshape
    """
    rows, cols = img.shape
    alist = (rows/n, n, cols/n, n)
    return img.reshape(alist).sum(1).sum(2) / (n*n)    

def rebinNP(img, n=2):
    """
    Using 2D array sum method
    """
    rows, cols = img.shape
    rn, cn = rows//n, cols//n
    out = np.zeros((rn,cn))
    for r in range(n):
        for c in range(n):
            out = out + img[r::n,c::n]
    return out / (n*n)
```

```python
bwImg = pilbw(orgImg)
```

```python
%%timeit
rebin2(bwImg)
```

```python
%%timeit
rebin2a(bwImg)
```

```python
%%timeit
rebinNP(bwImg)
```

## Now let's try Numba

```python
from numba import double, jit, i8

def rebinPurePy(mat, outMat, rows, cols, n=2):
    r = 0
    sum1 = 0
    c = 0
    r3 = 0
    c3 = 0
    r1 = 0
    c1 = 0
    r2 = 0
    c2 = 0
    while r1 < rows:
        c = 0
        c1 = 0
        #print ("r1", r1)
        while c1 < cols:
            sum1 = 0
            r2 = 0
            while r2 < n:
                r3 = r1 + r2
                c2 = 0
                while c2 < n:
                    c3 = c1 + c2
                    sum1 += mat[r3,c3]
                    c2 += 1
                r2 += 1
            outMat[r,c] = sum1 / n / n
            c += 1
            c1 += n
        r += 1
        r1 += n
    return outMat

locals1 = {'r':i8, 'c':i8, 'r1':i8, 'r2':i8}
rebinNumba = jit("i8[:,:](i8[:,:],i8[:,:],i8,i8,i8)", locals=locals1)(rebinPurePy)

height, width = bwImg.shape
out = np.zeros((height//2, width//2), dtype='double')
```

```python
%%timeit
a = rebinNumba(bwImg, out, height, width, 2)
```

## rebinNP is the fastest. This is surprising because I thought it needs more passes.

# Resampling

```python
def resample(a, n=2):
    """
    This is the version used in LACosmic, subsample()
    """
    newshape = (n*a.shape[0], n*a.shape[1])
    slices = [slice(0,old, float(old)/new) for old,new in zip(a.shape,newshape) ]
    coordinates = np.mgrid[slices]
    indices = coordinates.astype('i')   # choose the biggest smaller integer index
    return a[tuple(indices)]
```

```python
def resample2x2(img):
    """
    Using the Kronecker operator
    """
    m2 = np.ones((2,2))
    return np.kron(img, m2)
```

```python
%%timeit
resample(bwImg)
```

```python
%%timeit
resample2x2(bwImg)
```

### Using the Kronecker operator is three times faster.
