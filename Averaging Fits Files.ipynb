#How to calculate the average of FITS files?

For example: given a number of FITS files that represent biases of a detector, we want to calculate the average and use it as the master bias for further data reduction.

```python
import astropy.io.fits as pyfits
import matplotlib.pyplot as plt
import numpy as np
import math
import glob
```

```python
AllFitsFiles = glob.glob("../fits/esi1/e*.fits")
```

```python
def getHeader (fname):
    hdus = pyfits.open(fname)
    header = hdus[0].header
    hdus.close()
    return header;

def objectIs(fname, obj):
    header = getHeader(fname)
    return header["OBJECT"] == obj

Biases = [ f for f in AllFitsFiles if objectIs (f, "Bias") ]
```

```python
def getImageData (fname):
    hdus = pyfits.open(fname)
    data = hdus[0].data
    hdus.close()
    return data

def avgFits (fitsFiles):
    curr = getImageData(fitsFiles[0])
    for f in fitsFiles[1:]:
        curr = curr + getImageData(f)
    curr = curr / len (fitsFiles)
    return curr
        
```

```python
def show(data):
    fig = plt.figure(figsize=(15, 5))
    plt.subplot(111)
    im = plt.imshow(data, cmap="rainbow")
```

#Now call avgFits with the list of bias files

```python
# avgBias is the average of the biases
avgBias = avgFits(Biases)

hdr = getHeader(Biases[0])
masterBias = "../fits/esi1/masterBias.fits"
pyfits.writeto(masterBias, avgBias, header=hdr, clobber=True)
```

## Display the average

```python
# These next three functions are copied from Fits2Thumbnail.
# These are for displaying the result only, should not be modified.

def remap (arr, fromLo, fromHi, toLo, toHi):
    m = (toHi - toLo) / (fromHi - fromLo)
    b = -m * fromLo + toLo
    return np.clip(np.int_(arr * m + b), toLo, toHi)

def centroid(arr):
    l = arr.shape[0]
    ixs = np.arange(l)
    tmp = arr * ixs
    tmp1 = tmp * ixs
    sumarr = sum(arr)
    cen = sum(tmp)/sumarr
    return cen, math.sqrt(sum(tmp1)/sumarr - cen * cen)

def applyHEQ1(data, cutWidth=1, tFactor=1, nHist=65535):
    flatData = data.flatten()
    leng = len(flatData)
    histg, edges = np.histogram(flatData, bins=nHist, density=False)
    cen, cstd = centroid(histg)
    wing = cutWidth * cstd 
    loIdx = int(max(0, cen-wing))
    hiIdx = int(min(loIdx+wing+wing, nHist))
    fromLo = edges[loIdx]
    fromHi = edges[hiIdx]
    
    #print ("fromLo", fromLo, fromHi, cstd)
    data1 = remap(data, fromLo, fromHi, 0, nHist-1)
    histg, edges = np.histogram(data1.flatten(), bins=nHist, density=False)
    
    thold = leng / nHist * cstd * tFactor
    histg = np.clip(histg, 0, thold)
    hsum = np.cumsum(histg)
    data1 = hsum[np.int_(data1)]
    return data1

def data2Plot (subPlotId, data, **kwds):
    data1 = applyHEQ1(data, **kwds)
    ax = plt.subplot(subPlotId)
    ax.imshow(data1, cmap='gray', origin="lower")
    
fig = plt.figure(figsize=(14, 10))
data2Plot (111, getImageData (masterBias), tFactor=1)
```

While we are here, let's look for flats, average them and save it to "masterDomeFlat".

```python
DomeFlats = [ f for f in AllFitsFiles if objectIs (f, "Dome Flat") ]

hdr = getHeader(DomeFlats[0])
masterDomeFlat = "../fits/esi1/masterDomeFlat.fits"
pyfits.writeto(masterDomeFlat, avgFits(DomeFlats), header=hdr, clobber=True)
```

```python
# Let's see the master dome flat.
domeFlatData = getImageData(masterDomeFlat)
fig = plt.figure(figsize=(14, 10))
data2Plot (111, domeFlatData, tFactor=1)
```

```python
avgImag = avgFits(ObjFits)
```

```python
fig = plt.figure(figsize=(14, 10))
denom0 = domeFlatData - avgBias
denomSq = denom0 * denom0 + 1E-7
diff1 = (avgImag - avgBias ) * denom0 / denomSq
data2Plot (111, diff1, tFactor=1)
```
