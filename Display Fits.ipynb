# How to display a Fits image.

First the obligatory imports.

```python
%matplotlib inline
import pyfits
import numpy as np
import matplotlib.pyplot as plt
```

We simply read fits file and the get the image data from the first HDU.

```python
hdus = pyfits.open("abel254.fits")
data = hdus[0].data
```

The variable data now has the image data. It is a 300x300 pixel image. Its minimum and maximum pixel values can be easily obtained using the min() and max() functions.

```python
print (data.shape, "min=", data.min(), "max=", data.max())
```

Now we are ready to display the image.

First, we define the size of a figure. A figure can include multiple plots. In this example, we want to produce 4 plots in one single row. 

1. First plot shows the image using the default color map.
1. Second plot shows the image again, this time with a gray color map.
1. Third plot uses the reversed gray color map.
1. Four plot shows how to reverse the Y-axis.

```python
fig = plt.figure(figsize=(15, 5))
plt.subplot(141)
im = plt.imshow(data)
im.axes.set_xlabel("Default cmap")

plt.subplot(142)
im = plt.imshow(data, cmap="gray")
im.axes.set_xlabel("Gray color map")

plt.subplot(143)
im = plt.imshow(data, cmap="gray_r")
im.axes.set_xlabel("Reversed gray cmap")

plt.subplot(144)
im = plt.imshow(data, cmap="gray_r", origin='lower')
im.axes.set_xlabel("Y origin bottom left")
```

##More Plots

Matplotlib can plot the histogram of the pixel values with hist(). Note that the image must be flatten before calling hist() because data is a two-dimensional matrix and we want the histogram of all the pixels regardless where the pixels are.

Another way to plot a histogram is to call the function histogram(), which returns two arrays containing the histogram counts and the values of the bins, actually the values at the edges of the bins. There is one element in the bins array than in the histogram array. The function fill() is the actual plotting function. 

```python
fig = plt.figure(figsize=(13,4))
plt.subplot(121)
h = plt.hist(data.flatten(), bins=256)

ax = plt.subplot(122)
hist, bins = np.histogram(data.flatten(), bins=256, density=False)
ax.fill(bins[:-1], hist, "r", antialiased=True)
a = ax.set_xticks(range(2000, 20001, 2000))
```

Now, let's write a function that takes two numbers, the low cut and high cut values, the histogram and the bins as obatined before. The low cut and high cut values are numbers bewteen 0 and 1, where 0 means no pixels and 1 means all the pixels. The function should returns the pixel values corresponding to the low and high cut.  For example, a pixel value corresponding to a low cut of 0.1 means that 10% percent of the pixels have values less than that given pixel value.

```python
suma = np.cumsum(hist)
intensities = bins[:-1]
minVal, maxVal = suma[0], suma[-1]
dVal = maxVal - minVal
    
def getCuts(loCut, hiCut):
    loVal = dVal * loCut + minVal
    hiVal = dVal * hiCut + minVal
    minIdx = intensities[suma > loVal][0]
    maxIdx = intensities[suma < hiVal][-1]
    return (minIdx, maxIdx)
```

Using the function defined above, we create plots with different low and high cuts. 

This is probably not the best way to do plots. It is only an exercise.

```python
fig = plt.figure(figsize=(12, 4))

ax = plt.subplot(131)
minIdx, maxIdx = getCuts(0.1, 0.9)
im = ax.imshow(data, vmax=maxIdx, vmin=minIdx, cmap='gray_r')
ax.set_xlabel("0.1 to 0.9")

ax = plt.subplot(132)
minIdx, maxIdx = getCuts(0.5, 0.95)
im = ax.imshow(data, vmax=maxIdx, vmin=minIdx, cmap='gray_r')
ax.set_xlabel("0.5 to 0.95")

ax = plt.subplot(133)
minIdx, maxIdx = getCuts(0.01, 0.99)
im = ax.imshow(data, vmax=maxIdx, vmin=minIdx, cmap='gray_r')
ax.set_xlabel("0.01 to 0.99")
```

##Histogram Equalization

Histogram equalization is popular method to make details more visible. Here is an example of a traditional histogram equalized image plot with low and high cuts.

```python
def remap (arr, fromLo, fromHi, toLo, toHi):
    m = (toHi - toLo) / (fromHi - fromLo)
    b = -m * fromLo + toLo
    return np.clip(np.int_(arr * m + b), toLo, toHi)

def plotHistEq(data, loCut, hiCut, nHist=1024):
    histg, edges = np.histogram(data, bins=nHist, density=False)
    accSum = np.cumsum(histg)
    lo, hi = edges[0], edges[-1]
    fromLo = (hi-lo) * loCut + lo
    fromHi = (hi-lo) * hiCut + lo
    data1 = remap(data, fromLo, fromHi, 0, nHist-1)
    ax=plt.subplot(121)
    ax.plot (accSum, "b", label='acc sum')
    ax.legend(loc='lower right')
    ax1 = plt.twinx(ax)
    ax1.plot (histg, "r", label='histg')
    ax1.legend()
    data1 = accSum[data1]
    plt.subplot(122)
    plt.imshow(data1, cmap='gray_r')
    
```

```python
hdus = pyfits.open("abel254.fits")
data = hdus[0].data

fig = plt.figure(figsize=(10,4))
plotHistEq(data, 0.1, 0.99, 1024)
```

An simple enhancement of the traditional method, which can produce very noisy images, is the contrast limited histogram. 
Instead of accumulating the full amount of each bin in the histogram, a maximum limit is set. The result is a smoother image with lower contrast and less noisy.

```python
def plotHistEqA(data, loCut, hiCut, nHist=1024, tfactor=1):
    histg, edges = np.histogram(data, bins=nHist, density=False)
    accSum = [0] * nHist
    thold = 1.0 * len(data) / nHist * tfactor
    accSum[0] = histg[0]
    for i in range(1, nHist):
        inc = min(thold, histg[i])
        accSum[i] = inc + accSum[i-1]
    accSum = np.array(accSum)    
    lo, hi = edges[0], edges[-1]
    fromLo = (hi-lo) * loCut + lo
    fromHi = (hi-lo) * hiCut + lo
    data1 = remap(data, fromLo, fromHi, 0, nHist-1)
    ax = plt.subplot(121)
    ax.plot (accSum, label='histg')
    ax.legend(loc='lower right')
    
    data1 = accSum[data1]
    ax = plt.subplot(122)
    ax.imshow(data1, cmap='gray_r')
    return
```
